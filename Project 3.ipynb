### Q1
Steps to Calculate the Average Number of Daily Complaints in 2022:
- Filter for the Year 2022: Filter the DataFrame to include only entries from the year 2022.
- Resample Daily: Resample the filtered dataset on a daily basis, using the 'Unique Key' column to count the number of complaints each day. This column is used because it does not have missing values and each entry is unique to a specific complaint.
- Calculate the Average: Compute the average number of complaints from these daily counts.


```python
# Filter data for the year 2022
df_2022 = df[df.index.year == 2022]

# Resample the data to get daily counts of unique complaints
daily_complaints_2022 = df_2022['Unique Key'].resample('D').count()

# Calculate the average number of daily complaints
average_daily_complaints_2022 = daily_complaints_2022.mean()

# Print the result
average_daily_complaints_2022
```

### Q2

Steps to Find the Date with Maximum Calls:
- Resample Daily: Resample the data on a daily basis, using the 'Unique Key' column to count the number of complaints each day.
- Identify the Maximum: Identify the date with the highest number of daily complaints.


```python
# Resample the data to get daily counts of unique complaints
daily_complaints = df['Unique Key'].resample('D').count()

# Find the date with the maximum number of complaints
date_of_max_complaints = daily_complaints.idxmax()
max_complaints = daily_complaints.max()

# Print the result
print(f"The date with the maximum number of calls received is: {date_of_max_complaints.strftime('%Y-%m-%d')}")
print(f"Number of calls received on that date: {max_complaints}")
```

Q4

```python
# Group the data by month (ignoring the year), and count the number of complaints
monthly_complaints = df['Unique Key'].resample('ME').count()

# Aggregate the counts to get the total complaints for each month across all years
# 'M' gives end of month, so month number can be extracted like this:
total_monthly_complaints = monthly_complaints.groupby(monthly_complaints.index.month).sum()

# Find the month with the fewest calls
quietest_month = total_monthly_complaints.idxmin()

# Convert month number to month name
import calendar
quietest_month_name = calendar.month_abbr[quietest_month]

# Print the result
print(f"The quietest month historically, in terms of number of calls, is: {quietest_month_name}")
```

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Assuming df is already loaded and 'Created Date' is set as the datetime index
# If not, ensure to load and set it properly:
# df = pd.read_pickle('Project-3_NYC_311_Calls.pkl')
# df['Created Date'] = pd.to_datetime(df['Created Date'])
# df.set_index('Created Date', inplace=True)

# Check if the index is a DatetimeIndex
print(df.index)  # This will show the type of the index

# Resample the data to daily frequency, counting the number of complaints each day
daily_complaints = df['Unique Key'].resample('D').count()

# Perform ETS decomposition using an additive model
result = seasonal_decompose(daily_complaints, model='additive')

# Create a DataFrame with the decomposition results
ets = pd.DataFrame({
    'Total': daily_complaints,
    'Trend': result.trend,
    'Seasonality': result.seasonal,
    'Error': result.resid
})

# Display the first few entries of the DataFrame to check
print(ets.head(20))

# Filter the DataFrame to display only entries from December 2020 using proper slicing
december_2020_ets = ets['2020-12-01':'2020-12-31']  # Correct slicing by specifying the range

# Print the DataFrame for December 2020
print(december_2020_ets)

# Optionally, to focus on the seasonal component for December 25, 2020
seasonal_value_on_christmas = december_2020_ets.loc['2020-12-25', 'Seasonality']
rounded_seasonal_value = round(seasonal_value_on_christmas)
print(f"The seasonal component on 2020-12-25 is: {rounded_seasonal_value}")
```

```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_squared_error
from math import sqrt

# Load the data
df = pd.read_pickle('shared/Project-3_NYC_311_Calls.pkl')
df.index = pd.to_datetime(df.index)  # Ensure the index is in datetime format
df['count'] = 1  # Create a count column for aggregation

# Aggregate data to get daily counts
daily_df = df.resample('D').sum()

# Prepare data for Prophet
df_prophet = pd.DataFrame({
    'ds': daily_df.index,
    'y': daily_df['count']
})
```

```python
# Split data into training and testing sets
train_samples = len(df_prophet) - 90
train_set = df_prophet.iloc[:train_samples]
test_set = df_prophet.iloc[train_samples:]

# Create and fit the Prophet model
model = Prophet(daily_seasonality=True)
model.fit(train_set)

# Create future DataFrame for forecasting
future = model.make_future_dataframe(periods=90)

# Predict the future
forecast = model.predict(future)
```

```python
# Extract forecasted values for the test period
y_pred = forecast['yhat'].iloc[-90:].values
y_test = test_set['y'].values

# Calculate RMSE
rmse = sqrt(mean_squared_error(y_test, y_pred))
print(f'RMSE on the test set is: {rmse}')
```
